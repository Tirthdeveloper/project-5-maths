```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\DELL\Desktop\MATHS PROJECTS\PR-5\combined_employee_dataset_1.csv")
df
```

```python
employee1 = df.loc[0, ['age','years_experience','performance_score']].values
employee2 = df.loc[1, ['age','years_experience','performance_score']].values

print("Employee 1 Vector :", employee1)
print("Employee 2 Vector :", employee2)
```

```python
norm1 = np.linalg.norm(employee1, ord=1)

print("Norm-1 :", norm1)
```

```python
norm2 = np.linalg.norm(employee1)

print("Norm-2 :", norm2)
```

```python

dot_product = np.dot(employee1, employee2)

print("Dot Product :", dot_product)
```

```python

cos_theta = np.dot(employee1, employee2) / (np.linalg.norm(employee1) * np.linalg.norm(employee2))

angle = np.degrees(np.arccos(cos_theta))

print("Angle Between Vectors :", angle)
```

```python

employee1 = df.loc[0, ['age','years_experience','performance_score']].values
employee2 = df.loc[1, ['age','years_experience','performance_score']].values

v1 = employee1[:3]
v2 = employee2[:3]

cross_product = np.cross(v1, v2)

print("Cross Product :", cross_product)
```

```python

projection = (np.dot(employee1, employee2) / np.dot(employee2, employee2)) * employee2

print("Projection :", projection)
```

```python
from sklearn.decomposition import PCA

X = df[['age','years_experience','performance_score']]

pca = PCA(n_components=2)

pca_result = pca.fit_transform(X)

print(pca_result)
```

```python

pca_df = pd.DataFrame(pca_result, columns=['PC1','PC2'])

pca_df.head()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

X = df[['age','years_experience','performance_score']]

y = df['department']

lda = LinearDiscriminantAnalysis(n_components=1)

lda_result = lda.fit_transform(X, y)

print(lda_result)
```

```python

lda_df = pd.DataFrame(lda_result, columns=['LD1'])

lda_df.head()
```
